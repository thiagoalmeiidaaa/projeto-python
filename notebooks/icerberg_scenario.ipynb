```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("Iceberg Example") \
    .config("spark.sql.catalog.spark_catalog", "org.apache.iceberg.spark.SparkSessionCatalog") \
    .config("spark.sql.catalog.spark_catalog.type", "hive") \
    .getOrCreate()
```

```python
spark.sql("""
CREATE TABLE IF NOT EXISTS default.produtos (
    id INT,
    nome STRING,
    valor FLOAT
) USING iceberg
""")
```

```python
spark.sql("INSERT INTO default.produtos VALUES (1, 'Teclado', 120.0)")
spark.sql("INSERT INTO default.produtos VALUES (2, 'Mouse', 60.0)")
spark.sql("INSERT INTO default.produtos VALUES (3, 'Monitor', 800.0)")
```

```python
spark.sql("UPDATE default.produtos SET valor = 130.0 WHERE id = 1")
```

```python
spark.sql("DELETE FROM default.produtos WHERE id = 2")
```
